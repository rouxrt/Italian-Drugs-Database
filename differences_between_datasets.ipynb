```python
key = 'Codice AIC'
df_old.set_index(key, inplace=True)
df_new.set_index(key, inplace=True)

nuove = df_new[~df_new.index.isin(df_old.index)]

rimosse = df_old[~df_old.index.isin(df_new.index)]

comuni_old = df_old[df_old.index.isin(df_new.index)]
comuni_new = df_new[df_new.index.isin(df_old.index)]

modificate = comuni_new[comuni_new.ne(comuni_old)].dropna(how='all')

invariati = comuni_new[comuni_new.eq(comuni_old)].dropna(how='all')
```

```python
print(f"Added rows: {len(nuove)}")
print(f"Removed rows: {len(rimosse)}")
print(f"Modified rows: {len(modificate)}")
print(f"Unchanged rows: {len(comuni_new) - len(modificate)}")
```

```python
common_ids = df_old.index.intersection(df_new.index)
old_common = df_old.loc[common_ids]
new_common = df_new.loc[common_ids]

both_nan = old_common.isna() & new_common.isna()

diff_mask = (old_common != new_common) & ~both_nan

conteggio_modifiche = diff_mask.sum()

conteggio_modifiche = conteggio_modifiche[conteggio_modifiche > 0].sort_values(ascending=False)

print(conteggio_modifiche)

```

```python
print("Righe nuove:\n", nuove)
print("Righe rimosse:\n", rimosse)
print("Righe modificate:\n", modificate)
```

```python
differenze = []
colonne_comuni = df_old.columns.intersection(df_new.columns)

for idx in old_common.index:
    for col in colonne_comuni:
        old_val = old_common.loc[idx, col]
        new_val = new_common.loc[idx, col]

        if pd.isna(old_val) and pd.isna(new_val):
            continue

        if old_val != new_val:
            differenze.append({
                'Codice AIC': idx,
                'Campo': col,
                'Valore vecchio': old_val,
                'Valore nuovo': new_val
            })

df_diff = pd.DataFrame(differenze)

df_diff.dropna(how='all', inplace=True)


```

```python

# Trova righe nuove
nuove = df_new[~df_new.index.isin(df_old.index)]

# Trova righe rimosse
rimosse = df_old[~df_old.index.isin(df_new.index)]

# Trova righe comuni per confrontarle
comuni_old = df_old[df_old.index.isin(df_new.index)]
comuni_new = df_new[df_new.index.isin(df_old.index)]

# Trova righe modificate (differenze nei valori)
modificate = comuni_new[comuni_new.ne(comuni_old)].dropna(how='all')

# Righe invariate (opzionale)
invariati = comuni_new[comuni_new.eq(comuni_old)].dropna(how='all')
```

```python
print(f"Righe aggiunte:   {len(nuove)}")
print(f"Righe rimosse:    {len(rimosse)}")
print(f"Righe modificate: {len(modificate)}")
print(f"Righe invariate:  {len(comuni_new) - len(modificate)}")
```

```python


# Prendi solo le righe con id comuni
common_ids = df_old.index.intersection(df_new.index)
old_common = df_old.loc[common_ids]
new_common = df_new.loc[common_ids]

# Trova dove entrambi sono NaN
both_nan = old_common.isna() & new_common.isna()

# Trova le differenze ignorando i NaN uguali
diff_mask = (old_common != new_common) & ~both_nan

# Conta le modifiche per colonna
conteggio_modifiche = diff_mask.sum()

# Filtra solo i campi modificati e ordina
conteggio_modifiche = conteggio_modifiche[conteggio_modifiche > 0].sort_values(ascending=False)

print(conteggio_modifiche)

```

```python
differenze = []
colonne_comuni = df_old.columns.intersection(df_new.columns)

# Per ogni riga condivisa
for idx in old_common.index:
    for col in colonne_comuni:
        old_val = old_common.loc[idx, col]
        new_val = new_common.loc[idx, col]

        # Salta se entrambi sono NaN (quindi considerati "uguali")
        if pd.isna(old_val) and pd.isna(new_val):
            continue
        
        # Se sono diversi, registra la differenza
        if old_val != new_val:
            differenze.append({
                'Codice AIC': idx,
                'Campo': col,
                'Valore vecchio': old_val,
                'Valore nuovo': new_val
            })

# Crea il DataFrame delle differenze
df_diff = pd.DataFrame(differenze)

# (Opzionale) Rimuovi eventuali righe vuote
df_diff.dropna(how='all', inplace=True)

```
